```python
#defining a function to determine the position of zero element.
def zero_coord(node_i):
    for i in range(3):
        for j in range(3):
            if (node_i[i][j]==0):
                return i,j

Goal_state = ([1,2,3],
              [8,4,5],
              [7,6,0])
Row, Col=zero_coord(Goal_state)
print("Row "+ str(Row))
print("Column "+ str(Col))
```

```python
#defining function to swap the zero left
def left_slide(node_i, Row, Col):
    if Col==0:
        print("Cant shift left")
    else:
        swap = node_i[Row][Col-1]
        node_i[Row][Col-1]=0
        node_i[Row][Col]=swap
    return node_i
        
Current_state = ([1,2,3],
              [0,4,8],
              [7,6,5])
Row, Col=zero_coord(Current_state)
print(left_slide(Current_state, Row, Col))
```

```python
#defining function to swap the zero right
def right_slide(node_i, Row, Col):
    if Col==2:
        print("Cant shift right")
    else:
        swap = node_i[Row][Col+1]
        node_i[Row][Col+1]=0
        node_i[Row][Col]=swap
    return node_i
        
Current_state = ([1,2,3],
              [4,8,0],
              [7,6,5])
Row, Col=zero_coord(Current_state)
print(right_slide(Current_state, Row, Col))
```

```python
#defining function to swap the zero up
def up_slide(node_i, Row, Col):
    if Row==0:
        print("Cant shift up")
    else:
        swap = node_i[Row-1][Col]
        node_i[Row-1][Col]=0
        node_i[Row][Col]=swap
    return node_i
        
Current_state = ([1,0,3],
              [7,4,8],
              [2,6,5])
Row, Col=zero_coord(Current_state)
print(up_slide(Current_state, Row, Col))
```

```python
#defining function to swap the zero down
def down_slide(node_i, Row, Col):
    if Row==2:
        print("Cant shift down")
    else:
        swap = node_i[Row+1][Col]
        node_i[Row+1][Col]=0
        node_i[Row][Col]=swap
    return node_i
        
Current_state = ([1,0,3],
              [7,4,8],
              [2,6,5])
Row, Col=zero_coord(Current_state)
print(down_slide(Current_state, Row, Col))
```

```python
#checking if new node configuration already exists; if not appending the same onto the parent node:
def check_and_append(Parent_node, new_config, count):
    j=1
    for i in range(count):
        if (Parent_node[i]==new_config):  
            j=0
            break
        
    if j:
        Parent_node.append(new_config)
        count+=1
    return Parent_node, count

Count=3
P_node= [[[1,0,3],
          [7,4,8],
         [2,6,5]],
        [[1,2,3],
        [4,5,6],
        [7,0,8]],
        [[1,3,4],[2,5,6],[7,8,0]]]
n_config = [[1,2,3],[4,5,6],[7,8,0]]
P_node, Count = check_and_append(P_node,n_config,Count)
print(P_node)
print(Count)
```

```python
import copy

# Getting input from user
def zero_coord(node_i):
    for i in range(3):
        for j in range(3):
            if (node_i[i][j]==0):
                return i,j

#to check if goal is reached
def check_goal(Node_i, Goal_state):
    if Node_i==Goal_state:
        return 0
    else:
        return 1

#checking if new node configuration already exists; if not appending the same onto the parent node:
def check_and_append(Parent_node, new_config, count, Goal):
    temp_1=copy.deepcopy(Parent_node)
    c=count
    j=1
    Goal_num=0
    for i in range(c):
        if ((temp_1[i]==new_config) and j==1):
            j=0
    if j==1:
        temp_1.append(new_config)
        c+=1
    if (new_config==Goal):
        Goal_num=1
    return temp_1, c, Goal_num

#defining function to swap the zero left
def left_slide(node_i, Row, Col):
    temp = copy.deepcopy(node_i)
    if Col==0:
        print("Cant shift left")
    else:
        print("Shifting left")
        swap = temp[Row][Col-1]
        temp[Row][Col-1]=0
        temp[Row][Col]=swap
    return temp

#defining function to swap the zero right
def right_slide(node_i, Row, Col):
    temp = copy.deepcopy(node_i)
    if Col==2:
        print("Cant shift right")
    else:
        print("Shifting right")
        swap = temp[Row][Col+1]
        temp[Row][Col+1]=0
        temp[Row][Col]=swap
    return temp

#defining function to swap the zero up
def up_slide(node_i, Row, Col):
    temp = copy.deepcopy(node_i)
    if Row==0:
        print("Cant shift up")
    else:
        print("Shifting up")
        swap = temp[Row-1][Col]
        temp[Row-1][Col]=0
        temp[Row][Col]=swap
    return temp

#defining function to swap the zero down
def down_slide(node_i, Row, Col):
    temp = copy.deepcopy(node_i)
    if Row==2:
        print("Cant shift down")
    else:
        print("Shifting down")
        swap = temp[Row+1][Col]
        temp[Row+1][Col]=0
        temp[Row][Col]=swap
    return temp
            

Orig_P_node=[]
print("Enter entries row-wise")
Input_node = []
for i in range(3):    
    element_val = []
    for j in range(3):
        element_val.append(int(input()))
    Input_node.append(element_val)
print(Input_node)

Count = 1
Goal_state= [[1,2,3],
              [8,0,4],
              [7,6,5]]

Orig_P_node.append(Input_node)

G_count=0

inc=0
while(inc<Count):
    P_node= Orig_P_node[inc].copy()
    
    if not(G_count):
        Row, Col=zero_coord(P_node)
        temp1 = down_slide(P_node, Row, Col)
        Orig_P_node, Count, G_count = check_and_append(Orig_P_node, temp1, Count, Goal_state)
    else:
        print("Goal reached")
        break
    if not(G_count):
        Row, Col=zero_coord(P_node)
        temp1 = up_slide(P_node, Row, Col)
        Orig_P_node, Count, G_count = check_and_append(Orig_P_node, temp1, Count, Goal_state)
    else:
        print("Goal reached")
        break
    if not(G_count):
        Row, Col=zero_coord(P_node)
        temp1 = left_slide(P_node, Row, Col)
        Orig_P_node, Count, G_count = check_and_append(Orig_P_node, temp1, Count, Goal_state)
    else:
        print("Goal reached")
        break
    if not(G_count):
        Row, Col=zero_coord(P_node)
        temp1 = right_slide(P_node, Row, Col)
        Orig_P_node, Count, G_count = check_and_append(Orig_P_node, temp1, Count, Goal_state)
    else:
        print("Goal reached")
        break
    inc+=1

print("\n The unique combinations are as follows\n")
for lin in range(Count):
    for j in range(3):
        for k in range(3):
            print(Orig_P_node[lin][j][k], end=' ')
        print('')
    print('')
print(Count)
```
